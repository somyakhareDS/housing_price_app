```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train = pd.read_csv("train.csv", index_col = 'id')

train.head()
```

```python
for column in train.columns:
    print(column)
    print(train[column].value_counts().nlargest(5))
    print()
```

```python
len(train)
```

```python
train_only_sold = train[train['homeStatus'] == 'RECENTLY_SOLD'].copy()
train_only_sold
```

```python
train_only_sold.drop(["homeStatus","address","dateSold"], inplace = True, axis=1)
```

```python
train_only_sold
```

```python
numerical_columns = ['bathrooms','bedrooms','livingArea','lotArea','price' ]
```

```python
for column in numerical_columns:
    plt.title(column)
    plt.hist(train_only_sold[column])
    plt.show()
```

```python
square_feet_in_acre = 43560

def convert_lot_area(row):
    if row['lotUnit'] == 'acres':
        return row['lotArea'] * square_feet_in_acre
    else:
        return row.lotArea
    
train_only_sold.lotArea = train_only_sold.apply(convert_lot_area,axis=1)
```

```python
train_only_sold
```

```python
train_only_sold.drop('lotUnit',axis =1, inplace=True)
train_only_sold.head()
```

```python
bedroom_mean = train_only_sold.bedrooms.mean()
```

```python
bedroom_std_x_3 = train_only_sold.bedrooms.std()*3
```

```python
train_no_outliers = train_only_sold[np.abs(train_only_sold.bedrooms)-bedroom_mean<bedroom_std_x_3].copy()
```

```python
plt.hist(train_only_sold.bedrooms)
plt.show()

plt.hist(train_no_outliers.bedrooms)
plt.show()
```

```python
for column in numerical_columns:
    plt.title(column)
    plt.hist(train_no_outliers[column])
    plt.show()
```

```python
sns.heatmap(train_no_outliers.corr())
```

```python
median_income_by_zipcode = pd.read_csv('https://raw.githubusercontent.com/wlifferth/build-an-ml-web-app/main/median_income_by_zip_code.csv')
median_income_by_zipcode.head()
```

```python
merged_df = pd.merge(train_no_outliers, median_income_by_zipcode, how='left', left_on='zipcode', right_on='zip_code')
merged_df.head()
```

```python
merged_df.drop('zipcode',axis=1,inplace=True)
```

```python
merged_df.head()
```

```python
train_df = merged_df.copy()
```

```python
average_price = train_df['price'].mean()
average_price
```

```python
square_footage_model_df = train_df.copy()
average_price_per_square_foot = (square_footage_model_df['price']/square_footage_model_df['livingArea']).mean()
average_price_per_square_foot
```

```python
from sklearn.linear_model import LinearRegression

lin_reg_df = train_df.copy()

linear_regression_with_living_area_model = LinearRegression()

input_data = lin_reg_df[['livingArea']]
output_data = lin_reg_df['price']

linear_regression_with_living_area_model.fit(input_data, output_data)
```

```python
lin_reg_df['predicted'] = linear_regression_with_living_area_model.predict(input_data)
np.abs(lin_reg_df['predicted']-lin_reg_df['price']).mean()
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(lin_reg_df['predicted'],lin_reg_df['price'])
```

```python
lin_reg_df['livingAreaSquared'] = lin_reg_df['livingArea']**2
lin_reg_df['livingAreaRooted'] = lin_reg_df['livingArea']**0.5

input_data = lin_reg_df[['livingArea','livingAreaSquared','livingAreaRooted']]
output_data = lin_reg_df['price']

linear_regression_with_living_area_model.fit(input_data, output_data)

lin_reg_df['predicted'] = linear_regression_with_living_area_model.predict(input_data)
mean_absolute_error(lin_reg_df['predicted'],lin_reg_df['price'])
```

```python
columns = ['livingArea', 'livingAreaSquared', 'livingAreaRooted']
for i in range(2,5):
    column = f'livingAreaToThePowerOf{i}'
    columns.append(column)
    lin_reg_df[column] = lin_reg_df['livingArea'] ** i
    
input_data = lin_reg_df[columns]
output_data = lin_reg_df['price']

linear_regression_with_powers_model = LinearRegression()

linear_regression_with_powers_model.fit(input_data, output_data)

lin_reg_df['predicted'] = linear_regression_with_powers_model.predict(input_data)

mean_absolute_error(lin_reg_df.predicted, lin_reg_df.price)
```
